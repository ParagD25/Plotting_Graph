## Plotting Graph

```python
import pandas
from datetime import date
import matplotlib.pyplot as plt
data=pandas.read_excel('stock_data.xlsx')
data
```

```python
#data[(data['Date'].dt.date>=date(2020,7,1)) & (data['Date'].dt.date<=date(2020,12,31))]
```

### Taking Mean of entire Dataframe

```python
average=data.groupby(['Company']).mean()
average
```

### 'Close' Plotting of APPLE Stock 

```python
app=data[data['Company']=='APPLE']
plt.figure(figsize=(25,15))
plt.plot(app['Date'],app['Close'])
```

### Datewise averaging each stock 

```python
stock=data.groupby(['Date','Close','Company']).mean()
stock
```

### Closing price Above 1000

```python
plt.figure(figsize=(15,10))
plt.plot(data[data['Company']=='GOOGLE']['Date'],data[data['Company']=='GOOGLE']['Close'],label = "Google")
plt.plot(data[data['Company']=='RELIANCE']['Date'],data[data['Company']=='RELIANCE']['Close'],label = "Reliance")
plt.xlabel('Date -->')
plt.ylabel('Closing -->')
plt.title('Graphical Representation of Closing stock Price (above 1000)')
plt.legend()
```

### Closing Price above 100

```python
plt.figure(figsize=(15,10))
plt.plot(data[data['Company']=='APPLE']['Date'],data[data['Company']=='APPLE']['Close'],label="Apple")
plt.plot(data[data['Company']=='MICROSOFT']['Date'],data[data['Company']=='MICROSOFT']['Close'],label = "Microsoft")
plt.plot(data[data['Company']=='TATA MOTORS']['Date'],data[data['Company']=='TATA MOTORS']['Close'],label = "Tata Motors")
plt.xlabel('Date -->')
plt.ylabel('Closing -->')
plt.title('Graphical Representation of Closing stock Price (above 100)')
plt.legend()
```
